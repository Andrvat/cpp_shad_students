    ```c++

    std::string& r = name;
    r = "Dobrynia";
    std::cout << name << std::endl;

    std::string* p = &surname;
    surname = "Nikitich";
    std::cout << p << std::endl;
    ```

    Если у вас есть сложности с пониманием ссылок и указателей, рекомендуется проработать [статью](https://teccxx.neocities.org/mx1/pointers.html).


* Что следует сказать про эту программу?

```c++
std::vector<int> gen_sequence()
{
    return {1, 2, 3, 4, 5};
}
```

```c++
const int * search_value(const std::vector<int>& v, int value)
{
    for (int& x : v)
        if (x == value)
            return &x;
    return nullptr;
}

void f(int n)
{
    if (const int* val = find_value(gen_sequence(), n))
        std::cout << *val;
}
```
