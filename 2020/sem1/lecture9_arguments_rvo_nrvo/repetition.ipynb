```c++
std::vector<std::shared_ptr<Cat>> herd;
herd.reserve(100);
for (int i = 0; i < 100; ++i)
    herd.push_back(std::shared_ptr<Cat>(new Cat()));
```

```c++
std::vector<std::shared_ptr<Cat>> herd;
herd.reserve(100);
for (int i = 0; i < 100; ++i)
    herd.push_back(std::make_shared<Cat>());
```
