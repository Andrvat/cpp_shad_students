```c++
std::atomic<bool> is_published{false};
int data = 0;

void thread_1_worker()
{
    data = 1;
    is_published.store(true, std::memory_order_release);
}

void thread_2_worker()
{
    if (is_published.load(std::memory_order_acquire))
        assert(data == 1);
}
```
